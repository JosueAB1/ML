```python
## Importing libraries and reading the datatset "listings.csv"

import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from imblearn.over_sampling import SMOTE

df = pd.read_csv("listings.csv")
df
```

```python
# Dropping the dataset in variables 'X' and 'y'(Targets)
X = df.drop(["neighbourhood"], axis = 1)
y = df["neighbourhood"].values
```

```python
# Split 'X' and 'y' in train/test.

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
X_train.shape, X_test.shape, y_test.shape, y_test.shape
```

```python
X_train.columns
```

```python
# Removing the columns that aren't relevant to the analysis in 'X_train' and 'y_test'.

X_train = X_train.drop(["id", "name", "host_name", "neighbourhood_group", "last_review", "reviews_per_month", "license"], axis = 1)
X_test = X_test.drop(["id", "name", "host_name", "neighbourhood_group", "last_review", "reviews_per_month", "license"], axis = 1)
```

```python
X_train
```

```python
X_test
```

## Standarization

```python
# Standarizating the column 'room_type' from string to integer

encoder = LabelEncoder()
X_train["room_type"] = encoder.fit_transform(X_train["room_type"])
X_test["room_type"] = encoder.transform(X_test["room_type"])
```

## Scalling using MinManx

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

## LogisticRegression Model

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)
model
```

## SMOTE

```python
# Oversample minority class
sm = SMOTE(random_state=42)
X_train_oversampled, y_train_oversampled = sm.fit_resample(X_train, y_train)
```

# Training

```python
# Train model
model.fit(X_train_oversampled, y_train_oversampled)
score_test = model.score(X_test, y_test)
print("Score en test: ", score_test*100)
```

## Performance metrics

```python
from sklearn.metrics import classification_report

y_pred = model.predict(X_test)

# Calculate and print classification report
classification_rep = classification_report(y_test, y_pred, target_names=list(df["neighbourhood"].unique()))
print("Classification Report in Test:\n", classification_rep)
```
